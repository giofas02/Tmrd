3. Faccio delle nuove tabelle con i dati che mi interessano solamente

```python
coviddi1 = coviddi[['denominazione_regione', 'totale_positivi','isolamento_domiciliare']]
coviddi2 = coviddi[['denominazione_regione', 'totale_casi','dimessi_guariti']]
coviddi3 = coviddi[['denominazione_regione', 'deceduti']]


for i, (reg,valori) in enumerate(coviddi1.groupby('denominazione_regione')): #scorro le coppie (regione,valori) dove i valori sono tot positivi e isolamento
    #print(reg) - il groupby mi fa tante tabelle (una per regione, che è il primo indice, cioè reg) con i valori di coviddi1
    #print(valori) 
    if reg == 'Veneto': #se la reg è Veneto inseriscilo nella tabella di coppie (reg,valori)
        datiVeneto1 = valori.drop(['denominazione_regione'],axis = 1 ) #elimina la colonna con la regione
        
for (reg,valori) in (coviddi2.groupby('denominazione_regione')): #posso fare anche senza enumerate!
    if reg == 'Veneto':
        datiVeneto2 = valori.drop(['denominazione_regione'],axis = 1 )
        
for i, (reg,valori) in enumerate(coviddi3.groupby('denominazione_regione')):
    if reg == 'Veneto': 
        datiVeneto3 = valori.drop(['denominazione_regione'],axis = 1 )

print(datiVeneto1)
print(datiVeneto2)
print(datiVeneto3)
```

 4. Facciamo ora i grafici dei nuovi positivi del Veneto (asse y) per data (asse x)

```python
datiVeneto1.plot()
datiVeneto2.plot()
datiVeneto3.plot()
```
